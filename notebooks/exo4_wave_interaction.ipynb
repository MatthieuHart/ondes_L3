# Interaction between two waves

This notebook shows what happens when two waves cross each other in ideal (linear) conditions.

```python
# Basic imports

import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget
import copy

# Some useful functions and variables

def waveform_progressive(s):
    out = np.zeros(s.shape)
    pulse_indices = np.logical_and(s>-2, s<-1)
    out[pulse_indices] = 1
    return out

def waveform_retrograde(r):
    out = np.zeros(r.shape)
    pulse_indices = np.logical_and(r>1, r<3)
    out[pulse_indices] = 1
    return out

Ns = 1000
s_max = 5
s_vect = np.linspace(-s_max, s_max, Ns)


# Let's define a wavespeed
c = 1.5 # m.s-1

x = copy.deepcopy(s_vect) # our x-axis
```

# Linearity of the wave equation : the principle of superposition

We assume that a progressive wave $u_{+}(x,t)$ is coming from the left. At $t = 0$, we have 

$$
u_{+}(x,t = 0) = 1 \text{ if } x \in [-2,-1] \text{, } 0 \text{ else.}
$$

On the other hand, a retrograde wave $u_{-}(x,t)$ is coming from the right. At $t = 0$, we have 

$$
u_{-}(x,t = 0) = 1 \text{ if } x \in [1,3] \text{, } 0 \text{ else.}
$$

As the wave equation is linear, the resulting wave is the sum of both waves :

$$
u(x,t) = u_{+}(x,t) + u_{-}(x,t)
$$

## 1. Constructive interference

### At $t_0 = 0$ s

Let's apply this : the resulting shape at $t_0 = 0$ s is the sum of the propagative shape at $t_0$ (in orange on the figure) and of the retrograde shape at $t_0$ (in red on the figure). 

```python
# At time t0
t0 = 0 # s

# progressive wave 
s_t0 = x - c*t0 # phase of the progressive wave at t0
up_t0 = waveform_progressive(s_t0) # the shape of the progressive wave at t0

# progressive wave 
r_t0 = x + c*t0 # phase of the progressive wave at t0
ur_t0 = waveform_retrograde(r_t0) # the shape of the progressive wave at t0
print(ur_t0.shape)
# The resulting wave is just the sum
u_t0 = up_t0 + ur_t0

plt.figure(1)
plt.plot(x, u_t0, linewidth = 5)
plt.plot(x, up_t0,'-.', linewidth = 2)
plt.plot(x, ur_t0,'--', color = 'r', linewidth = 2)
plt.xlabel(r'$x$')
plt.ylabel(r'$u$')
plt.title('Waveform at $t_0 = 0$ s')
plt.legend(['$u = u_{+} + u_{-}$', '$u_{+}$', '$u_{-}$'])
plt.ylim([-3, 3])
plt.show()
```

### At $t_1 = 1$ s

And now let's look at what happens when the two waves cross each other, e.g. at $t_1 = 1$ s. Just as before, the resulting wave at $t_1$ is the sum of the propagative wave (in orange on the figure) and of the retrograde wave (in red on the figure). In this exercise, the waves interfere constructively when they cross each other.

```python
# At time t1
t1 = 1 # s

# progressive wave 
s_t1 = x - c*t1 # phase of the progressive wave at t0
up_t1 = waveform_progressive(s_t1) # the shape of the progressive wave at t0

# progressive wave 
r_t1 = x + c*t1 # phase of the progressive wave at t0
ur_t1 = waveform_retrograde(r_t1) # the shape of the progressive wave at t0

# The resulting wave is just the sum
u_t1 = up_t1 + ur_t1

plt.figure(2)
plt.plot(x, u_t1, linewidth = 5)
plt.plot(x, up_t1,'-.', linewidth = 2)
plt.plot(x, ur_t1,'--', color = 'r', linewidth = 2)
plt.xlabel(r'$x$')
plt.ylabel(r'$u$')
plt.title('Waveform at $t_1 = 1$ s')
plt.legend(['$u = u_{+} + u_{-}$', '$u_{+}$', '$u_{-}$'])
plt.ylim([-3, 3])
plt.show()
```

### At $t_2 = 2$ s

Same thing after they have crossed each other.

```python
# At time t2
t2 = 2 # s

# progressive wave 
s_t2 = x - c*t2 # phase of the progressive wave at t0
up_t2= waveform_progressive(s_t2) # the shape of the progressive wave at t0

# progressive wave 
r_t2 = x + c*t2 # phase of the progressive wave at t0
ur_t2 = waveform_retrograde(r_t2) # the shape of the progressive wave at t0

# The resulting wave is just the sum
u_t2 = up_t2 + ur_t2

plt.figure(3)
plt.plot(x, u_t2, linewidth = 5)
plt.plot(x, up_t2,'-.', linewidth = 2)
plt.plot(x, ur_t2,'--', color = 'r', linewidth = 2)
plt.xlabel(r'$x$')
plt.ylabel(r'$u$')
plt.title('Waveform at $t_2 = 2$ s')
plt.legend(['$u = u_{+} + u_{-}$', '$u_{+}$', '$u_{-}$'])
plt.ylim([-3, 3])
plt.show()
```
