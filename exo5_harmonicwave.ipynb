## Definition

The expression of a progressive harmonic wave is 

$$
u(x,t) = A cos [ 2 \pi (\dfrac{t}{T} - \dfrac{x}{\lambda}) ]
$$

The wave speed can be found starting from this idea : if I observe a phenomenon at point $x_0$ and time $t_0$, when will the same phenomenon be observed for the first time at $x_1 = x_0 + \Delta x$ ?

The last sentence translates to the following equation

$$
u(x_0, t_0) = u(x_1=0 + \Delta x, t_1),
$$

which is equivalent to 

$$
cos [ 2 \pi (\dfrac{t_0}{T} - \dfrac{x_0}{\lambda}) ] = cos [ 2 \pi (\dfrac{t_0 + \Delta t}{T} - \dfrac{x_0 + \Delta x}{\lambda}) ] = cos [ 2 \pi (\dfrac{t_0}{T} - \dfrac{x_0}{\lambda}) + 2 \pi  (\dfrac{\Delta t}{T} - \dfrac{\Delta x}{\lambda})].
$$

cos is $2 \pi$-periodic, so the first time the phenomenon appears again translates in maths to :

$$
2 \pi  (\dfrac{\Delta t}{T} - \dfrac{\Delta x}{\lambda}) = 2 \pi,
$$

which simply gives 

$$
\lambda = c T,
$$

which is a very important result for harmonic waves.

Note : 

We could also have written the wave as

$$
u(x,t) = A cos [ 2 \pi (\dfrac{t}{T} - \dfrac{x}{\lambda}) ] = A cos [ \dfrac{2\pi}{\lambda} (c t - x)],
$$

which is just at d'Alembert solution to the wave equation. Factorizing gives the same result as before : 

$$
\lambda = c T.
$$

```python
# Basic imports

import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget
import copy

# Some useful functions and variables
T = 1 # s, period of the wave
c = 1
lbd = c*T # this is what we've just shown

def waveform(x,t):
    out = np.cos(2*np.pi/lbd*(c*t - x))
    return out

# Let's define a wavespeed
Nx = 1000
x = np.linspace(0, 1, Nx)

t0 = T/5.7
T_values = [0, t0, T/4, T/2] #, 3*T/4, T]
plt.figure()
for T in T_values:
    u = waveform(x,T)
    plt.plot(x, u)
for T in T_values:
    plt.scatter(T, 0, marker = 'x', s = 100) #, c = 'r')
plt.legend(['t = 0', 'T/5.7', 'T/4'], loc = 'right')
plt.title('Propagation of an harmonic wave')

plt.grid()
plt.show()
```
